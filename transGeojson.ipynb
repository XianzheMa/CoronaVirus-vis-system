```python
import os
import json
import pandas as pd
```

```python
# load renaming dictionary
_CHN_EN_DICT_ = './data/locationDict.csv'
chn_en = pd.read_csv(_CHN_EN_DICT_, encoding='utf-8')
bad_frm = chn_en[chn_en['Chinese'].str.contains('市|省')]
rename_dict = {name: name.split('市')[0] if '市' in name else name.split('省')[0] for name in bad_frm['Chinese']}
rename_dict['内蒙古自治区'] = '内蒙古'
rename_dict['新疆维吾尔自治区'] = '新疆'
rename_dict['宁夏回族自治区'] = '宁夏'
rename_dict['广西壮族自治区'] = '广西'
rename_dict['西藏藏族自治区'] = '西藏'
rename_dict['西藏自治区'] = '西藏'
chn_en['Chinese'] = chn_en['Chinese'].replace(rename_dict)

rename_dict = dict(zip(chn_en['Chinese'], chn_en['English']))

with open('./data/geojson/china.geo.json') as f:
    geojson = json.load(f)

for feature in geojson['features']:
    old_name = feature['properties']['name']
    if old_name in rename_dict:
        feature['properties']['name'] = rename_dict[old_name]
    else:
        print(old_name + ' does not exist in rename_dict.')

with open('./data/geojson/new/China.geo.json', 'w') as f:
    json.dump(geojson, f)
```

```python
# load fileAccessor.json
with open('./data/geojson/fileAccessor.json') as f:
    fileAccessor = json.load(f)
```

```python
prefix = './data/geojson/'
new_prefix = './data/geojson/new/'
for key, value in fileAccessor.items():
    oldFileName = prefix + key + '.geo.json'
    
    with open(oldFileName, 'r') as f:
        geojson = json.load(f)
    
    for feature in geojson['features']:
        old_name = feature['properties']['name']
        if old_name[-1] == '市' or (len(old_name) > 2 and old_name[-1] == '县'):
            old_name = old_name[0:-1] # drop '市', '县'
        if old_name in rename_dict:
            feature['properties']['name'] = rename_dict[old_name]
        else:
            print(old_name + ' does not exist in rename_dict for {}.'.format(value))
    
    newFileName = new_prefix + rename_dict[value] + '.geo.json'
    with open(newFileName, 'w', encoding='utf-8') as f:
        json.dump(geojson, f, ensure_ascii=False)
```
